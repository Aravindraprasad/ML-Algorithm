### Ordinal Encoder
Ordinal data should be ordered, such as Poor > Avg > Good.  

- Applied for X columns [inputs]

- Applies to data labels like - Poor, Avg, Good.

- This should only be applied to X columns [inputs].

### Label Encoder
Label data should be ordered, such as Poor > Avg > Good.

- Applied for Y columns [outputs]

- Applies to data labels like - Poor, Avg, Good.

- This should only be applied to Y columns [outputs].

```python
# Here review	education are ordinal Data

df = pd.read_csv('/content/customer.csv')
df.head()
```

```python
new_df = df.iloc[:,2:]
```

```python
new_df.head(10)
```

```python
# Seperated the dataset into X & Y
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(new_df.iloc[:,0:2],new_df.iloc[:,-1],test_size=0.2)
X_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
