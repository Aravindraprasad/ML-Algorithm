```python
from sklearn.datasets import make_regression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='Model')
plt.legend()
plt.show()
```

```python
# Building linear model using Gradient Descent

class MeraLR:
  def __init__(self,learning_rate, epochs):
    self.b = 0
    self.m = 1
    self.learning_rate = learning_rate
    self.epochs = epochs

  def fit(self,X,y):
    for i in range(self.epochs):
      loss_slope_b = -2 * np.sum(y - self.m * X.ravel() - self.b)
      loss_slope_m = -2 * np.sum((y - self.m * X.ravel() - self.b) * X.ravel())

      self.b = self.b - (self.learning_rate * loss_slope_b)
      self.m = self.m - (self.learning_rate * loss_slope_m)

    print(self.m, self.b)

  def predict(self, X):
    return (self.m * X + self.b)

```

```python
learning_rate = 0.001
epochs = 100
gd = MeraLR(learning_rate, epochs)
```

```python
gd.fit(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X,gd.predict(X),color='red')
```
