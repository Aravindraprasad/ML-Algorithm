###One Hot Encoding

This is encoding which is used to convert Nominal data into Numerical Data

Nominal Data Ex-Red, green, Blue,Yellow, which cannot be ranked!

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/cars.csv')
```

### Nominal Data
- brand -> 32 brands
- fuel -> Diesel Petrol CNG	LPG

- owner	-> First Owner ,Second Owner,Third Owner	,Fourth & Above Owner,
Test Drive Car

```python
df['owner'].value_counts()
```

1. OneHotEncoding using Pandas

```python
pd.get_dummies(df,columns=['fuel','owner'])
```

 ### 2. K-1 OneHotEncoding :

- First Column is removed
  
- This is done to remove Multiple-colinerity which create issue in Linear Regression algorithm

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

### 4. OneHotEncoding with Top Categories

```python

counts = df['brand'].value_counts()
```

```python

df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python

pd.get_dummies(df['brand'].replace(repl, 'uncommon')).sample(5)
```
